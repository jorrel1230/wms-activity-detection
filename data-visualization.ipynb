# Data Visualization

- The main purpose of this file is to get a handle on what the time-series features look like.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wisdm_path = 'scratch/network/jr8867/wisdm-dataset'
processed_watch_path = f'/{wisdm_path}/processed'

df = pd.read_csv(os.path.join(processed_watch_path, 'processed_watch_data.csv'))
```

### What we gather from raw sensor vs time graphs

There are drastically different patterns from the different activity types, over time. This is likely good data to use in time series models such as RNNs and LSTMs.

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate mean and variance for each activity
mean_accel_A = subject_0_activity_A[['ax', 'ay', 'az']].mean()
variance_accel_A = subject_0_activity_A[['ax', 'ay', 'az']].var()

mean_accel_B = subject_0_activity_B[['ax', 'ay', 'az']].mean()
variance_accel_B = subject_0_activity_B[['ax', 'ay', 'az']].var()

mean_accel_D = subject_0_activity_D[['ax', 'ay', 'az']].mean()
variance_accel_D = subject_0_activity_D[['ax', 'ay', 'az']].var()

mean_accel_E = subject_0_activity_E[['ax', 'ay', 'az']].mean()
variance_accel_E = subject_0_activity_E[['ax', 'ay', 'az']].var()

# Create a range of values for plotting
x = np.linspace(-15, 20, 100)

# Create subplots for each axis
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

# Standard colors for activities
colors = {
    'Walking': 'blue',
    'Jogging': 'orange',
    'Sitting': 'green',
    'Standing': 'red'
}

# Plot Gaussian distributions for ax
for activity_mean, activity_var, activity_name in zip(
    [mean_accel_A, mean_accel_B, mean_accel_D, mean_accel_E],
    [variance_accel_A, variance_accel_B, variance_accel_D, variance_accel_E],
    ['Walking', 'Jogging', 'Sitting', 'Standing']
):
    axs[0].plot(x, (1/np.sqrt(2 * np.pi * activity_var['ax'])) * np.exp(-0.5 * ((x - activity_mean['ax'])**2) / activity_var['ax']), color=colors[activity_name], label=f'{activity_name} (ax)')

# Plot Gaussian distributions for ay
for activity_mean, activity_var, activity_name in zip(
    [mean_accel_A, mean_accel_B, mean_accel_D, mean_accel_E],
    [variance_accel_A, variance_accel_B, variance_accel_D, variance_accel_E],
    ['Walking', 'Jogging', 'Sitting', 'Standing']
):
    axs[1].plot(x, (1/np.sqrt(2 * np.pi * activity_var['ay'])) * np.exp(-0.5 * ((x - activity_mean['ay'])**2) / activity_var['ay']), color=colors[activity_name], label=f'{activity_name} (ay)')

# Plot Gaussian distributions for az
for activity_mean, activity_var, activity_name in zip(
    [mean_accel_A, mean_accel_B, mean_accel_D, mean_accel_E],
    [variance_accel_A, variance_accel_B, variance_accel_D, variance_accel_E],
    ['Walking', 'Jogging', 'Sitting', 'Standing']
):
    axs[2].plot(x, (1/np.sqrt(2 * np.pi * activity_var['az'])) * np.exp(-0.5 * ((x - activity_mean['az'])**2) / activity_var['az']), color=colors[activity_name], label=f'{activity_name} (az)')

# Set titles and labels
axs[0].set_title('Gaussian Distributions for ax')
axs[0].set_xlabel('Acceleration Values')
axs[0].set_ylabel('Probability Density')
axs[0].legend()
axs[0].grid()

axs[1].set_title('Gaussian Distributions for ay')
axs[1].set_xlabel('Acceleration Values')
axs[1].set_ylabel('Probability Density')
axs[1].legend()
axs[1].grid()

axs[2].set_title('Gaussian Distributions for az')
axs[2].set_xlabel('Acceleration Values')
axs[2].set_ylabel('Probability Density')
axs[2].legend()
axs[2].grid()

plt.tight_layout()
plt.show()
```

### What we gather from Overall Mean and Variances

Although means aren't that significant, the variance is clearly extremely important in the detection of activity, as is seems to have a direct relationship with how much the activity "makes you move". This leads us to a key insight - we probably **shouldn't** normalize our features to have mean 0 and variance 1 - variance is precisely what gives us knowledge on the activity.

Feeding in features such as variances can be very useful in activity detection, which could be very useful in traditional models like random forests.

```python
# Extract the first row of segment_features_0_A
first_row = segment_features_0_A.iloc[0]

# Prepare data for plotting
data = {
    'ax_mean': first_row['ax_mean'],
    'ay_mean': first_row['ay_mean'],
    'az_mean': first_row['az_mean'],
    'gx_mean': first_row['gx_mean'],
    'gy_mean': first_row['gy_mean'],
    'gz_mean': first_row['gz_mean'],
    'ax_var': first_row['ax_var'],
    'ay_var': first_row['ay_var'],
    'az_var': first_row['az_var'],
    'gx_var': first_row['gx_var'],
    'gy_var': first_row['gy_var'],
    'gz_var': first_row['gz_var']
}
print(data)

# Create figures for accelerations and gyros
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 15))

# Create Gaussian distributions for accelerations (ax, ay, az) in green, red, and blue
for key_mean, key_var, color in zip(['ax_mean', 'ay_mean', 'az_mean'], 
                                     ['ax_var', 'ay_var', 'az_var'], 
                                     ['green', 'red', 'blue']):
    mean = data[key_mean]
    variance = data[key_var]
    x = np.linspace(mean - 4 * np.sqrt(variance), mean + 4 * np.sqrt(variance), 100)
    y = (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-0.5 * ((x - mean) ** 2) / variance)
    
    ax1.plot(x, y, label=f'Gaussian for {key_mean[:-5]}', color=color)

ax1.set_title('Gaussian Distributions for Accelerations')
ax1.set_xlabel('Values')
ax1.set_ylabel('Probability Density')
ax1.legend()
ax1.grid()

# Create Gaussian distributions for gyros (gx, gy, gz) in blue, red, and green
for key_mean, key_var, color in zip(['gx_mean', 'gy_mean', 'gz_mean'], 
                                     ['gx_var', 'gy_var', 'gz_var'], 
                                     ['blue', 'red', 'green']):
    mean = data[key_mean]
    variance = data[key_var]
    x = np.linspace(mean - 4 * np.sqrt(variance), mean + 4 * np.sqrt(variance), 100)
    y = (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-0.5 * ((x - mean) ** 2) / variance)
    
    ax2.plot(x, y, label=f'Gaussian for {key_mean[:-5]}', color=color)

ax2.set_title('Gaussian Distributions for Gyros')
ax2.set_xlabel('Values')
ax2.set_ylabel('Probability Density')
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```
